```python
df['wd2'] = df['wd']
```

```python
for i in df.index:
    if df.wd[i] >= 30 and df.wd[i] < 120:
        df.wd2[i] = df.wd[i] - 30
    if df.wd[i] >= 120 and df.wd[i] < 210:
        df.wd2[i] = df.wd[i] - 120
    if df.wd[i] >= 210 and df.wd[i] < 300:
        df.wd2[i] = df.wd[i] - 210
    if df.wd[i] >= 300 and df.wd[i] < 360:
        df.wd2[i] = df.wd[i] - 300
    if df.wd[i] >= 0 and df.wd[i] < 30:
        df.wd2[i] = abs(df.wd[i] - 30)
```

```python
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
reg = linear_model.Ridge(alpha = 0.5)
reg2 = linear_model.Ridge(alpha = 0.5)
reg3 = linear_model.BayesianRidge()
poly = PolynomialFeatures(degree=2)
```

```python
reg.fit (df[['wd','ws']], df['wp1']) 
reg2.fit (df[['wd2','ws']], df['wp1']) 
reg3.fit (df[['wd2','ws']], df['wp1']) 
X = poly.fit_transform(df[['wd2','ws']])
reg.fit(X, df['wp1'])
```

```python
from sklearn.metrics import mean_squared_error

print('wd:%f' %mean_squared_error(df['wp1'], reg.predict(df[['wd','ws']])))
print('wd2:%f' %mean_squared_error(df['wp1'], reg2.predict(df[['wd2','ws']])))
print('wd2:%f' %mean_squared_error(df['wp1'], reg3.predict(df[['wd2','ws']])))
```
